# Bidirectional Configuration Plots

## Imports

```python
# dependencies
import numpy as np
import os 
import sys

# qom modules
from qom.ui import init_log
from qom.ui.plotters import MPLPlotter

# add path to local libraries
sys.path.append(os.path.abspath(os.path.join('../..', 'sync_bi_uni')))
# import system
from systems.Bi00 import Bi00

# initialize logging
init_log()
```

## Constants and Variables

```python
# frequently used variables
file_path = 'H:/Workspace/data/bi_00'

# variables
deltas = np.around(np.linspace(-0.02, 0.02, 101), 4).tolist()
lambs = np.around(np.linspace(0.0, 0.1, 101), 3).tolist()
X, Y = np.meshgrid(deltas, lambs)
file_suffix = '_1e3-20pi_52.0_1.0_0.005_[0.005, 0.005]_[0.005, 0.005]_[0.15, 0.15]_0.075_[0, 0]_1.0_delta_-0.02_0.02_101_lamb_0.0_0.1_101.npz'
```

## Phonon Number Difference

```python
# get differences
n_b_diff = np.load(file_path + '/n_b_diff' + file_suffix)['arr_0'].tolist()
n_b_diff_nan = [[n_b_diff[i][j] if np.abs(n_b_diff[i][j]) <= 500 else np.NaN for j in range(len(n_b_diff[i]))] for i in range(len(n_b_diff))]

# plotter
plotter = MPLPlotter({
    'X': deltas,
    'Y': lambs
}, {
    'type': 'contourf',
    'palette': 'RdBu_r',
    'bins': 51,
    'x_label': '$\\delta / \\omega_{mL}$',
    'x_bound': 'both',
    'x_ticks': [-0.02, 0.00, 0.02],
    'y_label': '$\\lambda / \\kappa$',
    'y_bound': 'both',
    'y_ticks': [0.0, 0.0375, 0.075],
    'y_tick_labels': [0.0, 0.25, 0.5],
    'show_cbar': True,
    'cbar_ticks': [-10.0, 0.0, 10],
    'cbar_tick_labels': ['(-)ve', 'zero', '(+)ve'],
    'cbar_position': 'right',
    'label_font_size': 22,
    'tick_font_size': 18
})
plotter.update(xs=X, ys=Y, vs=n_b_diff_nan)
plotter.show(True, 6.0, 5.0)
```

## Average Phase Synchronization

```python
# get phase synchronization
S_phase_avg = np.load(file_path + '/S_phase_avg' + file_suffix)['arr_0']
S_phase_avg = [[S_phase_avg[i][j] if n_b_diff_nan[i][j] == n_b_diff_nan[i][j] else 0 for j in range(len(n_b_diff_nan[i]))] for i in range(len(n_b_diff_nan))]

# plotter
plotter = MPLPlotter({
    'X': deltas,
    'Y': lambs
}, {
    'type': 'contourf',
    'palette': 'RdBu_r',
    'bins': 11,
    'x_label': '$\\delta / \\omega_{mL}$',
    'x_bound': 'both',
    'x_ticks': [-0.02, 0.0, 0.02],
    'y_label': '$\\lambda / \\kappa$',
    'y_bound': 'both',
    'y_ticks': [0.0, 0.0375, 0.075],
    'y_tick_labels': [0.0, 0.25, 0.5],
    'cbar_ticks': [0.0, 0.1, 0.2],
    'label_font_size': 22,
    'tick_font_size': 18
})
plotter.update(xs=X, ys=Y, vs=S_phase_avg)
plotter.show(True, 6.0, 5.0)
```

## Gaussian Quantum Discord

```python
# get Gaussian discord
D_G_avg = np.load(file_path + '/D_Gaussian_avg' + file_suffix)['arr_0']
D_G_avg = [[D_G_avg[i][j] if n_b_diff_nan[i][j] == n_b_diff_nan[i][j] else 0 for j in range(len(n_b_diff_nan[i]))] for i in range(len(n_b_diff_nan))]

# plotter
plotter = MPLPlotter({
    'X': deltas,
    'Y': lambs
}, {
    'type': 'contourf',
    'palette': 'Reds',
    'bins': 11,
    'x_label': '$\\delta / \\omega_{mL}$',
    'x_bound': 'both',
    'x_ticks': [-0.02, 0.0, 0.02],
    'y_label': '$\\lambda / \\kappa$',
    'y_bound': 'both',
    'y_ticks': [0.0, 0.0375, 0.075],
    'y_tick_labels': [0.0, 0.25, 0.5],
    'cbar_title': '$\\langle D_{G} \\rangle$',
    'cbar_ticks': [0.00, 0.01, 0.02],
    'cbar_position': 'top',
    'label_font_size': 22,
    'tick_font_size': 18
})
plotter.update(xs=X, ys=Y, vs=D_G_avg)
plotter.show(True)
```
